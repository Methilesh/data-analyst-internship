```python
import pandas as pd
import sqlite3
con = sqlite3.connect('/content/sql-murder-mystery.db')
```

```python
query1 = '''
SELECT *
FROM sqlite_master
WHERE type = 'table'
'''
pd.read_sql_query(query1, con)
```

```python
#Checking schema of crime scene report table
query2 = '''
SELECT sql
FROM sqlite_master
WHERE name = 'crime_scene_report'
'''
pd.read_sql_query(query2, con)
```

```python
pd.set_option('display.max_colwidth', None)
#Bringing up the crime scene report
query3 = '''
SELECT *
FROM crime_scene_report
WHERE city = 'SQL City'
ORDER BY date
'''
pd.read_sql_query(query3, con)
```

```python
query4 = '''
SELECT sql
FROM sqlite_schema
WHERE name = 'person';
'''
pd.read_sql_query(query4, con)
#Bringing up name of witness
query5 = '''
SELECT *
FROM person
WHERE address_street_name = "Northwestern Dr"
ORDER BY address_number DESC
LIMIT 1;
'''
pd.read_sql_query(query5, con)
```
